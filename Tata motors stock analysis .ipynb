```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
dates = pd.date_range(start='2025-02-1', end='2025-03-16', freq='B')  # Business days
open_prices = np.random.uniform(low=100, high=500, size=len(dates))  # Random opening prices

# Create a DataFrame
df = pd.DataFrame({'Date': dates, 'close': open_prices})
df.set_index('Date', inplace=True)

# Plot stock price trends
plt.figure(figsize=(12, 6))
plt.plot(df.index, df["close"], label="Opening Price", color="blue")  # Use the correct column name
plt.title("Tata Motors Stock Price (2024-2025)")
plt.xlabel("Year")
plt.ylabel("Stock Price (INR)")
plt.legend()
plt.show()
```

```python
# Plot moving average (50-day)
df["50-day MA"] = df["close"].rolling(window=50).mean()
plt.figure(figsize=(12, 6))
plt.plot(df["close"], label="closing Price", color="blue")
plt.plot(df["50-day MA"], label="50-Day Moving Average", color="red")
plt.title("Tata Motors Stock Price with Moving Average")
plt.xlabel("Year")
plt.ylabel("Stock Price (INR)")
plt.legend()
plt.show()
```
